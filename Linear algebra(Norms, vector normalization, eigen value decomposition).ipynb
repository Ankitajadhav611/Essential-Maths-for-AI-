```python
import numpy as np
```

```python
A = np.arange(9)-3
A
```

```python
B = A.reshape((3,3))
B
```

```python
# Norm is the magnitude of an vector, distance between the origin and the vector
# to implement norm import form numpy linear algebra module 
from numpy import linalg
```

```python
# this is the default L2 norm also know as euclidean norm
A_norm = np.linalg.norm(A)
A_norm
```

```python
# L1 norm
A_norm1 = np.linalg.norm(A,1)
A_norm1
```

```python
# max norm also know as with l is infinity 
A_normi = np.linalg.norm(A,np.inf)
A_normi
```

vector normalisation, also know as unit value vector generation, where in convert each member value of an vector in range of 0 to 1, this is essential step of data preparation for an neural network model.

```python
# normalisation is diving the vector with its norm 
norm = np.linalg.norm(A)
A_unit = A/norm
A_unit
```

```python
# to verify the magnitude of an unit vector
np.linalg.norm(A_unit)            
```

Eigenvalues & Eigenvector:
for a given matrix the multiple of a eigenvector results in the product of scalar multiple of eigenvector times the eigenvalues
   

A x lambda = lambda x vec

```python
vec1 = np .arange(1,10)
A = vec1.reshape(3,3)
eigen_value , eigen_vector = np.linalg.eig(A)
```

```python
eigen_value
```

```python
eigen_vector
```

the eigenvalue[1] corresponds to eigenvector[:,1] , to aprove the dependency 

```python
u = eigen_vector[:,1]
```

```python
lam = eigen_value[1]
```

```python
np.dot(A,u)
```

```python
lam*u
```

Eigen Decompostion :
    To verify our calculation further and extract the original matrix from calculated eigenvalues and eigenvector

```python
matrix = np.matmul(np.diag(eigen_value),np.linalg.inv(eigen_vector))
result = np.matmul(eigen_vector,matrix).astype(np.int)
result
```

we were able to retrive the original matrix thus it worked well
